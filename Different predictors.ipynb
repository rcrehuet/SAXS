# Check how different predictors give different curves

We try the following predictors:

* crysol
* FOXS
* AXES
* aquaSAXS
* Waxsis

In all cases we used the same PDB (pcnaN.pdb) and the default options for the codes/servers.

## Read outputs

```python
import numpy as np
import scipy.interpolate
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
crysol = np.loadtxt('pcnaN00.int')[:,:2]
crysol = crysol[crysol[:,0]<=0.5]
```

```python
foxs = np.loadtxt('pcnaN.pdb-FOXS.dat')[:,:2]
foxs = foxs[foxs[:,0]<=0.5]
```

```python
axes = np.loadtxt('AXES/pcnaN.dat')[:,:2]
axes = axes[axes[:,0]<=0.5]
```

```python
aquasaxs =np.loadtxt('aquasaxs/pcna.saxs')[:,:2]
aquasaxs = aquasaxs[aquasaxs[:,0]<=0.5]
```

```python
waxsis = np.loadtxt('waxsresults.1410650/intensity.dat')[:, :2]
waxsis = waxsis[waxsis[:,0]<=0.5]
```

```python
exp = np.loadtxt('curve_exp2.dat')
exp = exp[exp[:,0]<=0.5]
```

## define function to calculate $\chi^2$ and scaled curves

```python
def calc_chi(data, dataref):
        """
        Get the chi for the data relative to the experimental rataref
        with error standard deviation error_sd.
        Return chi for scaled curve and scaled curve with constant substraction. Return also the curve
        """
        ref_q = dataref[:, 0]
        ref_y = dataref[:, 1]
        sd = dataref[:, 2]

        # Keep the ref_q range within the calculated q range
        valid_range = np.where((ref_q>data[0,0]) & (ref_q<data[-1,0]))
        ref_q = ref_q[valid_range]
        ref_y = ref_y[valid_range]
        sd = sd[valid_range]
 
        ## Reduce the number of points of data, but keep first and last for the interp1d
        ind= np.zeros_like(data[:,0], dtype=bool)
        #ind[::5]=True
        #ind[-1]=True
        # KEEP ALL POINTS
        ind[:] = True
        fit = scipy.interpolate.interp1d(data[ind,0],  data[ind, 1], kind='cubic')
        calc_y = fit(ref_q)
 
        # scale the calculated and experimental curve
        sum1 = np.sum(calc_y*ref_y/sd**2)
        sum2 = np.sum(calc_y*calc_y/sd**2)
        ratio = sum1/sum2
        
        #find scaling based on the low q range
        lowqInd = ref_q<0.05
        sum1 = np.sum(calc_y[lowqInd]*ref_y[lowqInd]/sd[lowqInd]**2)
        sum2 = np.sum(calc_y[lowqInd]*calc_y[lowqInd]/sd[lowqInd]**2)
        lowqRatio = sum1/sum2        
        
        calc_y *= ratio
        # Calculate chi square        
        chi = np.sum((ref_y-calc_y)**2/sd**2)/(len(ref_q)-1)
        # Linear fit calculated data to experimental data
        sol = np.polyfit(calc_y, ref_y, 1, w = 1/sd)
        calc_y_cst = sol[0]*calc_y+sol[1]
        chi_cst = np.sum((ref_y-calc_y_cst)**2/sd**2)/(len(ref_q)-1)

        calc_y *= lowqRatio/ratio
        # Calculate chi square        
        lowqchi = np.sum((ref_y-calc_y)**2/sd**2)/(len(ref_q)-1)
        # Linear fit calculated data to experimental data
        sol = np.polyfit(calc_y, ref_y, 1, w = 1/sd)
        calc_y_cst = sol[0]*calc_y+sol[1]
        lowqchi_cst = np.sum((ref_y-calc_y_cst)**2/sd**2)/(len(ref_q)-1)

        
        #Save the data in a text file
        #np.savetxt('saxy3_fits.dat', np.c_[ref_q, ref_y, calc_y, calc_y_cst])
        return (chi, lowqchi), chi_cst, ref_q,(calc_y*ratio/lowqRatio, calc_y)
```

## Plot results

First we call our function to get the scaled curves and the $\chi^2$.

```python
crysolChi2, chi2cst, q,crysolScaled = calc_chi(crysol, exp)
foxsChi2, chi2cst, q,foxsScaled= calc_chi(foxs, exp)
axesChi2, chi2cst, qaxes,axesScaled = calc_chi(axes, exp)
aquasaxsChi2, chi2cst, q,aquasaxsScaled = calc_chi(aquasaxs, exp)
waxsisChi2, chi2cst, qwaxsis,waxsisScaled = calc_chi(waxsis, exp)
```

Now we plot the curves, scaled to minimize the $\chi^2$ along the complete $q$ range, i.e. up to $q=0.5$. 

```python
plt.figure(figsize=(14,9))
plt.errorbar(exp[:,0], exp[:, 1], yerr=exp[:, 2], errorevery=5, alpha=0.7, color='r')
plt.semilogy(q, crysolScaled[0], 'g-', label='Crysol')
plt.semilogy(q, foxsScaled[0], 'b-', label='FOXS')
plt.semilogy(qaxes, axesScaled[0], 'm-', label='AXES')
plt.semilogy(q, aquasaxsScaled[0], 'c-', label='aquaSAXS')
plt.semilogy(qwaxsis, waxsisScaled[0], 'k-', label='WAXSIS')
plt.xlim((0.0, 0.5))
plt.legend(loc='best');
```

Now the curves scaled only in the region of $q<0.05$.

```python
plt.figure(figsize=(14,9))
plt.errorbar(exp[:,0], exp[:, 1], yerr=exp[:, 2], errorevery=5, alpha=0.7, color='r')
plt.semilogy(q, crysolScaled[1], 'g-', label='Crysol')
plt.semilogy(q, foxsScaled[1], 'b-', label='FOXS')
plt.semilogy(qaxes, axesScaled[1], 'm-', label='AXES')
plt.semilogy(q, aquasaxsScaled[1], 'c-', label='aquaSAXS')
plt.semilogy(qwaxsis, waxsisScaled[1], 'k-', label='WAXSIS')

plt.xlim((0.0, 0.5))
plt.legend(loc='best');
```

Here are the $\chi^2$ for the two types of scaling. Remark that $\chi^2$ is calculated in both cases for $q<0.5$. Therefore it is always larger for scaling of $q<0.05$. However it is interesing that the trends are not conserved.

WAXIS is always the best, with difference! Crysol is always the second. However aquaSAXS, which was the worst now comes the third. And AXES, which was performing close to Crysol is now the worst. 

These differences are hard to see in the log curve. I wonder if one should also calculate $\chi^2$ for the log of the intensity, and not for the intensity as we quantify agreements based on $\chi^2$ calculated for the raw intensity.
